## Question 2 (Easy)

**Title:** Sum of First N Numbers

### Problem Statement:
Take an integer N as input and find the sum of numbers from 1 to N.

### Approach:
Initialize sum = 0, loop from 1 to N, add each value to sum.


```python
N = int(input())
sum = 0

for i in range(1, N):
    sum += i

print(sum)
```

## Question 9 (Easy)

**Title:** Count Even and Odd Numbers

### Problem Statement:
Take N as input and count how many even and odd numbers are there between 1 and N.

Use two counters and check with modulo operator.


```python
N = int(input())
even = 0
odd = 0
for i in range(1, N+1):
    if i % 2 == 0:
        even += 1
    else:
        odd += 1

print(even, odd)
```

## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
N = int(input())

if N <= 1:
    print("Not Prime")
else:
    for i in range(2, N-1):
        if N % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
```
