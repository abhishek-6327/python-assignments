### Hard 2

**Question:**  
Find the second largest element.

**Approach:**  
Track largest and second largest values.


```python
numbers = [10, 45, 23, 89, 34]

largest = second_largest = float('-inf')

for num in numbers:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num != largest:
        second_largest = num

print(second_largest)
```

### Hard 3

**Question:**  
Rotate a list to the right by k steps.

**Approach:**  
Use slicing and modulo operation.


```python
numbers = [1, 2, 3, 4, 5]
k = 2

k = k % len(numbers)

rotated_list = numbers[-k:] + numbers[:-k]
print(rotated_list)
```

### Hard 5

**Question:**  
Split a list into even and odd lists.

**Approach:**  
Check each element and append accordingly.


```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

even = []
odd = []

for num in numbers:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)

print("Even list:", even)
print("Odd list:", odd)
```

### Hard 6

**Question:**  
Find common elements between two lists.

**Approach:**  
Use loop or set intersection.


```python
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7]

common = []
for item in list1:
    if item in list2 and item not in common:
        common.append(item)

print(common)
```

### Hard 8

**Question:**  
Find frequency of each element.

**Approach:**  
Use dictionary to count occurrences.


```python
numbers = [1, 2, 2, 3, 3, 3, 4]

freq = {}

for num in numbers:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

print(freq)
```

### Hard 9

**Question:**  
Move all zeros to end of list.

**Approach:**  
Create two lists or swap in-place.


```python
numbers = [0, 1, 0, 3, 12]

non_zero = []
zeros = []

for num in numbers:
    if num == 0:
        zeros.append(num)
    else:
        non_zero.append(num)

result = non_zero + zeros
print(result)
```

### Hard 10

**Question:**

**Approach:**  
Compare current and previous values.


```python
numbers = [1, 3, 2, 4, 4, 6, 5]

count = 0
for i in range(1, len(numbers)):
    if numbers[i] > numbers[i - 1]:
        count += 1

print(count)
```
