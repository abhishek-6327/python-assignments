# Map, Filter, Reduce

## Problem 1: Square all numbers using `map()`

**Approach:**
- Take a list of numbers
- Use `map()` to apply square operation on each element
- Convert result to list

```python
numbers = [1, 2, 3, 4, 5]

squared_numbers = list(map(lambda x: x * x, numbers))

print(squared_numbers)
```

## Problem 2: Convert list of strings to uppercase

**Approach:**
- Use `map()`
- Apply `upper()` on each string

```python
words = ["apple", "banana", "cherry"]

uppercase_words = list(map(lambda x: x.upper(), words))

print(uppercase_words)
```

## Problem 3: Filter even numbers using `filter()`

**Approach:**
- Use `filter()`
- Condition: number % 2 == 0

```python
numbers = [1, 2, 3, 4, 5, 6]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print(even_numbers)
```

## Problem 4: Filter names longer than 4 characters

**Approach:**
- Use `filter()`
- Keep strings with length > 4

```python
names = ["Amit", "Rahul", "Sneha", "Karan", "Priyanka"]

long_names = list(filter(lambda x: len(x) > 4, names))

print(long_names)
```

## Problem 5: Find sum of numbers using `reduce()`

**Approach:**
- Import reduce from functools
- Add elements step-by-step

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

total_sum = reduce(lambda x, y: x + y, numbers)

print(total_sum)
```

## Problem 6: Find product of numbers

**Approach:**
- Use reduce
- Multiply elements

```python
from functools import reduce

numbers = [1, 2, 3, 4]

product = reduce(lambda x, y: x * y, numbers)

print(product)
```

## Problem 7: Combined Map + Filter

**Problem:** Square numbers and keep only even results

**Approach:**
- Map to square
- Filter even values

```python
numbers = [1, 2, 3, 4, 5]

squared = map(lambda x: x * x, numbers)
even_squares = list(filter(lambda x: x % 2 == 0, squared))

print(even_squares)
```
