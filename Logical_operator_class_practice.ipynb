## Problem 2: Leap Year

**Question:** Check if a given year is a leap year.

**Approach:** Apply the rules: divisible by 4, but centuries must be divisible by 400.

Rules: Year divisible by 4 is leap, but if divisible by 100 it must also be divisible by 400.

```python
# Enter your answer here
year= int(input("enter a year: "))
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("leap year")

        else:
            print("not a leap year")
    else:
        print("leap year")

else:
    print("not a leap year")
```

## Problem 5: Movie Ticket Discount

**Question:** Apply discount based on age.

**Approach:** Use ordered elif statements from smallest to largest age group.

Apply discounts based on age:
- Age < 12 → 50% discount
- Age between 12 and 18 → 30% discount
- Age between 18 and 60 → no discount
- Age > 60 → 40% discount

```python
age = int(input("Enter your age: "))

if age < 12:
    discount = 0.5
elif age <= 18:
    discount = 0.3
elif age <= 60:
    discount = 0
else:
    discount = 0.4
print(f"Discount: {discount}%")
```

## Problem 7: Delivery Charge System

**Question:** Calculate final delivery charge based on prime membership, amount, and distance.

**Approach:** Start with prime-check, then inside it apply nested if blocks.

Inputs:
- order_amount (int)
- is_prime_member (yes/no)
- distance_km (float)

Rules:
- If prime member → Free delivery
- Else:
  - If amount ≥ 500:
    - If distance ≤ 5 km → Free
    - Else → ₹20 extra
  - Else:
    - If distance ≤ 5 km → ₹30
    - Else → ₹50

```python
order_amount = int(input("Enter order amount: "))
is_prime_member = input("Are you a prime member? (yes/no): ")
distance_km = float(input("Enter delivery distance in km: "))

if is_prime_member == 'yes':
  delivery_charge = 0
else:
  if order_amount >= 500:
    if distance_km <= 5:
      delivery_charge = 0
    else:
      delivery_charge = 20
  else:
    if distance_km <= 5:
      delivery_charge = 30
    else:
      delivery_charge = 50
print(f"Delivery Charge: ₹{delivery_charge}")

```

## Problem 8: ATM Withdrawal Simulation

**Question:** Simulate ATM rules: multiples of 100, sufficient balance, manager approval.

**Approach:** Check errors first, then nested conditions for amount > 10000.

Inputs:
- balance
- amount

Logic:
- If amount % 100 != 0 → reject (ATM works with ₹100 multiples)
- If amount > balance → reject
- Else:
  - if amount > 10000 → ask for manager approval
  - else → withdraw success

```python
balance = float(input("Enter your account balance: "))
amount = float(input("Enter the amount to withdraw: "))

if amount % 100 != 0:
  print("Rejected: Amount must be in multiples of 100")
elif amount > balance:
  print("Rejected: Insufficient balance")
else:
  if amount > 10000:
    print("Withdrawal requires manager approval")
  else:
    print("Withdrawal successful")
```

## Problem 9: Employee Performance Rating

**Question:** Assign performance rating based on sales, feedback, attendance.

**Approach:** Prioritize attendance rule, then apply nested conditions.

Inputs:
- sales (₹)
- feedback (0-5 scale)
- attendance (%)

Logic:
- If attendance < 70 → 'Poor'
- Else:
  - If sales > 100000 AND feedback > 4.5 → 'Excellent'
  - If sales > 70000 OR feedback > 4 → 'Good'
  - Else → 'Average'

```python
sales = float(input("Enter sales amount (₹): "))
feedback = float(input("Enter feedback score (0-5): "))
attendance = float(input("Enter attendance percentage: "))
if attendance < 70:
  rating = 'Poor'
else:
  if sales > 100000 and feedback > 4.5:
    rating = 'Excellent'
  elif sales > 70000 or feedback > 4:
    rating = 'Good'
  else:
    rating = 'Average'
print(f"Performance Rating: {rating}")

```
