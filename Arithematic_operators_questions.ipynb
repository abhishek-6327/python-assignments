## 1. Calculate the Total Bill
A restaurant bill is ₹450. GST is 18%. Calculate the final bill amount.

```python
# Write your solution here
bill=450

gst=0.18*bill
sgst=0.12*bill
other=5

final_bill=bill+gst+sgst+other

print(final_bill)
```

## 3. Temperature Converter
Convert 36°C to Fahrenheit using F = (C * 9/5) + 32.

```python
# Write your solution here

temp= 36
Fahrenheit= (temp* 9/5) + 32

print(Fahrenheit)
```
